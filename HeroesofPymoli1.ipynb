```python
import pandas as pd
import numpy as np
import os
```

```python
file = os.path.join("..", "pandas-challenge", "purchase_data.csv") 
```

```python
purchase_data = pd.read_csv(file)
purchase_data
```

```python
#Total Number of Players 
total_players = purchase_data["SN"].unique()
total_players
```

```python
total_players = pd.DataFrame(total_players).count()
```

```python
totalplayers = pd.DataFrame(total_players, columns = ['Total Players'])
totalplayers

```

```python
#Unique Items
unique_items = purchase_data["Item Name"].unique()
uniqueitems = pd.DataFrame(unique_items).count()
uniqueitems
```

```python
#Average Price 
average_price = purchase_data["Price"].mean()
average_price
```

```python
#Total purchases 
total_purchase = purchase_data["Price"].sum()
total_purchase
```

```python
Summary_table = pd.DataFrame({'Average Price': average_price,
                    'Number of Purchases': total_purchase,
                    'Number of Unique Items': uniqueitems,
                    'Total Revenue': total_purchase})
Summary_table
```

```python
#Demographics Gender 
gender_stats = purchase_data.groupby("Gender")
total_count_gender = gender_stats.nunique()["SN"]
total_count_gender

#gender_demographics = pd.DataFrame({"Percentage of Players": percentage_of_players, "Total Count": total_count_gender})
#gender_demographics.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}"})
```

```python
#Percentage of Players
percentage_of_players = total_count_gender / float(total_players) *100
percentage_of_players
```

```python
gender_demographics = pd.DataFrame({"Percentage of Players": percentage_of_players, "Total Count": total_count_gender})
gender_demographics.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}"})
```

```python
#Age Demographics 
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],bins, labels=group_names)
purchase_data
```

```python
age_grouped = purchase_data.groupby("Age Group")
age_grouped
```

```python
total_count_Edad = age_grouped["SN"].nunique()
total_count_Edad
```

```python
percentage_by_age = (total_count_Edad/float(total_players)) * 100
percentage_by_age
```

```python
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": total_count_Edad})
age_demographics
```

```python
#Style 
age_demographics.style.format({"Percentage of Players":"{:,.2f}"})
```

```python
#Purchasing Analysis 
purchase_count_age = age_grouped["Purchase ID"].count()
```

```python
avg_purchase_price_age = age_grouped["Price"].mean()
avg_purchase_price_age
```

```python
total_purchase_value = age_grouped["Price"].sum()
total_purchase_value
```

```python
avg_purchase_per_person_age = total_purchase_value/total_count_Edad
avg_purchase_per_person_age
```

```python
Summary_age_demo = pd.DataFrame({"Purchase Count": purchase_count_age,
                                 "Average Purchase Price": avg_purchase_price_age,
                                 "Total Purchase Value":total_purchase_value,
                                 "Average Purchase Total per Person": avg_purchase_per_person_age})
Summary_age_demo
```

```python
Summary_age_demo.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

```python
#Top Spenders 
spender_group = purchase_data.groupby("SN")
```

```python
spender_group_count = spender_group["Purchase ID"].count()
spender_group_count
```

```python
avg_spender_group = spender_group["Price"].mean()
avg_spender_group
```

```python
spender_group_total = spender_group["Price"].sum()
spender_group_total
```

```python
top_spenders_table = pd.DataFrame({"Purchase Count": spender_group_count,
                             "Average Purchase Price":avg_spender_group ,
                             "Total Purchase Value":spender_group_total})
top_spenders_table
```

```python
top_spenders_table.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```
